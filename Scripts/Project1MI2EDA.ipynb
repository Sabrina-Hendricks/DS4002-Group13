```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Import data from github

```python
url = 'https://raw.githubusercontent.com/Sabrina-Hendricks/DS4002-Group13/main/Womens%20Clothing%20E-Commerce%20Reviews.csv'
df = pd.read_csv(url)
df.head()
```

Graph 1: Histogram of the number of reviews per product type

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Class Name', order=df['Class Name'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Reviews per Product Type')
plt.xlabel('Product Type')
plt.ylabel('Number of Reviews')
plt.show()
```

Graph 2: Age distribution of reviewers

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=15, kde=True)
plt.title('Distribution of Age of Reviewers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
